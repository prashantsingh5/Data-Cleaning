# Changing the Datatypes of columns

```python
data["BLOCK"] = data["BLOCK"].astype(float)
data["LOT"] = data["LOT"].astype(float)
data[" ZIP CODE"] = data[" ZIP CODE"].astype(float)
data["RESIDENTIAL UNITS"] = data["RESIDENTIAL UNITS"].astype(float)
data["COMMERCIAL UNITS"] = data["COMMERCIAL UNITS"].astype(float)
data["TOTAL UNITS"] = data["TOTAL UNITS"].astype(float)
data["LAND SQUARE FEET"] = data["LAND SQUARE FEET"].astype(float)
data["GROSS SQUARE FEET"] = data["GROSS SQUARE FEET"].astype(float)
data["TAX CLASS AT TIME OF SALE"] = data["TAX CLASS AT TIME OF SALE"].astype(float)
data["SALE PRICE"] = data["SALE PRICE"].astype(float)    # In case of model training this column acts as target variable
data['SALE DATE'] = data['SALE DATE'].str.split(' ').str[0]   # removing time from this column because time is always zero
```
